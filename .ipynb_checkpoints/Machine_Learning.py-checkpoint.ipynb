```python
#Required lib
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
from sklearn.model_selection import train_test_split



#read the data_set
df=pd.read_csv("Downloads//test.csv")

#clean up the data_set
df['Rent'] = df['Rent'].replace(['-'],'0')
df['Rent']=pd.to_numeric(df['Rent'])
df['MaintenanceCharge'] = df['MaintenanceCharge'].replace(['No'],'0')
df['MaintenanceCharge']=pd.to_numeric(df['MaintenanceCharge'])
df[["Area","Measure"]]= df["Area"].str.split(" ", n = 0, expand = True)
df['Area'] = df['Area'].str.replace(',','')
df['Area']=pd.to_numeric(df['Area'])
df=df[df['Place']!='-']
dff=df[['Place','Area','MaintenanceCharge','Deposit','Rent','Furnished','ApartmentType','TenantType']]




#Analysing the data_set
Amounts=['Rent','Deposit','MaintenanceCharge']
for Amount in Amounts:
    dfA=dff[['Place',Amount]]
    result = dfA.groupby(['Place']).mean().plot.bar()

for Amount in Amounts:
    dfA=dff[['Place',Amount]]
    result1 = dfA.groupby(['Place']).max().plot.bar()
    
for Amount in Amounts:
    dfA=dff[['Place',Amount]]
    result2 = dfA.groupby(['Place']).min().plot.bar()

dfp=dff[['Place','ApartmentType']]
result3 = dfp.groupby(['Place','ApartmentType'])['ApartmentType'].count().plot.bar()



#check the outlayer and remove those
plt.boxplot(dff['Deposit'])
plt.show()
plt.boxplot(dff['Rent'])
plt.show()
plt.boxplot(dff['MaintenanceCharge'])
plt.show()
data_set=dff[(dff['Rent']<=80000) & (dff['MaintenanceCharge']<=400000) & (dff['Deposit']<=200000)]


#Check the corelation and reduce the corelation
data_set['Maintance']=np.sin(data_set['MaintenanceCharge'])
data_set['Rent_']=np.sin(data_set['Rent'])
corrMatrix = data_set[['Rent_','Deposit','Maintance']].corr() 
sns.heatmap(corrMatrix, annot=True, cmap="viridis") 
plt.show() 

#dummy variable
data_set=pd.get_dummies(data_set)
data_set=data_set.drop(['Rent','MaintenanceCharge'],axis=1)


#scaleup the data_set
scaledX = scale.fit_transform(data_set)


#test and train the data_set
X=data_set.drop(['Rent_'], axis=1)
y=data_set['Rent_']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)



#predict the data_set using multiple regression
regr = linear_model.LinearRegression()
regr.fit(X, y)


predicted = regr.predict([[880,65000,0.428264,0,0,1,0,1,0,0,0,1,0,0,0,0,0,1]])

```
